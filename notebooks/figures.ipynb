```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
from cheopslcs import LCSim
```

```python
s1 = LCSim(window_width=5.0, exp_time=60, white_noise=1e-4)
s2 = LCSim(window_width=24*1.4, exp_time=60, white_noise=1e-5)
```

```python
time1, flux1 = s1(radius_ratio=0.1, zero_epoch=0.0, period=2.0,
                 scaled_semi_major_axis=8.0, impact_parameter=0.1,
                limb_darkening=[0.2, 0.3])

time2, flux2 = s2(radius_ratio=0.1, zero_epoch=0.0, period=0.7,
                 scaled_semi_major_axis=4.0, impact_parameter=0.1,
                 eccentricity=0.2, argument_of_periastron=0.23*3.14,
                 geometric_albedo=0.5, limb_darkening=[0.2, 0.3],
                 efficiency=0.6, eff_phase=0.2)

fig, axs = plt.subplots(1, 2, figsize=(10, 3), constrained_layout=True)
axs[0].plot(time1, flux1, '.')
axs[1].plot(time2, flux2, '-')
plt.setp(axs, xlabel='Time [d]')
plt.setp(axs[0], ylabel='Flux', title='Simple transit');
plt.setp(axs[1], ylim=(0.9995, 1.0005), title='Phase curve');
fig.savefig('example.svg')
```
